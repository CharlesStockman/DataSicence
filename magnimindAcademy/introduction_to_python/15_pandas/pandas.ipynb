## Create a crosstab 

between state and category showing amount pledged as the value. 

Please refer to the following link for this practice:

A pivot table can also be used.  The difference between a crosstab and pivot table is A pivot spreadsheet is a broader definition similar to cross tab. While the cross tab assumes there is only one horizontal dimension and all the values belong to that dimension, pivot assumes there might exist one or more horizontal dimensions.

```python
data_frame_11 = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/ksprojects.csv')
data_frame_11
```

```python
# 
pd.crosstab(index=data_frame_11.state, columns=data_frame_11.category,values=data_frame_11['pledged'],\
            aggfunc=sum)
```

```python
import numpy as np
pd.set_option('Display.max_columns', None, 'Display.max_rows', None)
res_1 = pd.pivot_table(data_frame_11,index='state', columns='category', values='pledged', aggfunc=[sum, np.mean, np.std])
res_1
```
