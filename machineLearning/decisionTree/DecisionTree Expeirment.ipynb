```python
# Setup the imports
import seaborn as sns
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
```

## Load the Data Sets into memory

```python
# Load the Titanic Data Set
titanic_dataset = sns.load_dataset('titanic')
print("Shape of dataset -- Titanic = ", titanic_dataset.shape)
titanic_dataset
```

```python
# Load the ISIS Dataset
isis_dataset = sns.load_dataset('iris')
print("Shape of dataset Iris = ", isis_dataset.shape)
isis_dataset
```

## Analyze the Data
Since I am using decision trees.  I found out an interesting fact that classical decision  cannot deal with 
categorical data, so a solution is to use DataFrame.getdummies() which creates one-hot encoding

### Find and Remove all Columns with NA
Since this is an experiment,  I just want enoght preprocessing to provide great accuracy and to study data to prove/disprove my hypothesis

```python
titanic_dataset = titanic_dataset.dropna(axis=1, how='any' )
#
titanic_dataset
```

```python
## Study which rows can be one encoded since Decision Tree can' handle categorical data
```

```python
# Commented out the column(s) that had NA and were removed in the 2nd previous block above.

print("Showing the columns that can be converted to one hot encoding")
print("*** pclass")
print(titanic_dataset["pclass"].value_counts())
print("*** Sex")
print(titanic_dataset['sex'].value_counts())
print("*** Sibling and Spouses")
print(titanic_dataset['sibsp'].value_counts())
print("*** Number of Parents/Children Aboard")
print(titanic_dataset['parch'].value_counts())
#print("*** Embarked Town")
#titanic_dataset['embark_town'].value_counts()
print("***Class")
print(titanic_dataset['class'].value_counts())
print("*** alone")
print(titanic_dataset['alone'].value_counts())
```

```python
data_frame_tmp = pd.get_dummies(titanic_dataset['pclass'], prefix='pclass_')
titanic_dataset =pd.concat([titanic_dataset, data_frame_tmp], axis=1)
titanic_dataset =titanic_dataset.drop('pclass', axis=1)

data_frame_tmp = pd.get_dummies(titanic_dataset['sex'], prefix='sex_')
titanic_dataset = pd.concat([titanic_dataset, data_frame_tmp], axis=1)
titanic_dataset = titanic_dataset.drop('sex', axis=1)

data_frame_tmp = pd.get_dummies(titanic_dataset['sibsp'], prefix="sibsp_")
titanic_dataset = pd.concat([titanic_dataset, data_frame_tmp], axis=1)
titanic_dataset = titanic_dataset.drop('sibsp', axis=1)

data_frame_tmp = pd.get_dummies(titanic_dataset['parch'], prefix='parch')
titanic_dataset = pd.concat([titanic_dataset, data_frame_tmp], axis=1)
titanic_dataset = titanic_dataset.drop('parch', axis=1)

data_frame_tmp = pd.get_dummies(titanic_dataset['alone'], prefix='alone_')
titanic_dataset = pd.concat([titanic_dataset, data_frame_tmp], axis=1)
titanic_dataset = titanic_dataset.drop('alone', axis=1)

titanic_dataset
```

## Create the Testing and Training sets

```python
#
# Create a collection of training and test sets  
def create_testing_training_sets(data):
   sets =  [ train_test_split(data, train_size = index) \
            for index in np.linspace(start=.1, stop=.9,num=9).tolist()]
   return sets;

def display_test_train_counts(data_set_name="None Provided", data=[]):
    print("Data Set: ", data_set_name)
    for index in range(0,9):
        print ("\t", index, len(data[index][0]),len(data[index][1]))   

titanic_test_training_sets = create_training_sets(titanic_dataset)
display_test_train_counts("titanic", titanic_test_training_sets)

isis_test_training_sets = create_testing_training_sets(isis_dataset)
display_test_train_counts("isis", isis_test_training_sets)

```
