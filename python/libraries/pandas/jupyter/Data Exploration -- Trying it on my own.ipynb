```python
data_frame_1 = pd.read_csv('gdp_per_capita_2020July.csv', skiprows=4)
```

```python
# Get a list of columns
data_frame_1.columns
```

```python
# Verify that the indicator code has only one column -- If it does then we can make a constant and decrease the 
# table frame.
data_frame_1["Indicator Code"].value_counts()
```

```python
# Get the ammount of memory that the for the current data frame.
data_frame_1.info(verbose=False, memory_usage='deep')
```

```python
# The column has the same value for each rows: Indicator Code and Indicator Name.  We can get rid of one column
# and descrease the number of columns.
print(data_frame_1['Indicator Code'].value_counts())
print(data_frame_1['Indicator Name'].value_counts())
data_frame_2 = data_frame_1.drop(['Indicator Name', 'Indicator Code'], axis=1)
data_frame_2.info(memory_usage='deep', verbose=False)
```

```python
# We don't need 2 Columns ( Country Name, Country ) to identify the count.  We can create a function to get the
# country name from the country code.  There the country name can be put into its own function.
data_frame_3 = data_frame_2.drop('Country Name', axis=1)
data_frame_3.info(memory_usage='deep', verbose = False)
```

```python
# Show the data frame 
data_frame_3
```

```python
# Get the columns which we will not be working with and drop them from data frame.
canidates=[]
for i in range(1960, 2017):
    canidates.append(str(i))
canidates.append('Unnamed: 64')
print("Canidates to be removed:", canidates)
data_frame_4 = data_frame_3.drop(canidates, axis=1)
```

```python
# Here we have the dataframe with columns we will work with
data_frame_4
```

```python
# Show the amount of memory that we are currently using
data_frame_4.info(memory_usage='deep')
```

```python
pd.isnull(data_frame_4['2018']).to_numpy().nonzero()
```

```python
# fill in missing values by using linear interopolation.  Based on my understanding of economics it does change
# to widely.  Later on we can study economic output and create a better distruction.
#
# Since the data in the first column did not have any previous data we filled it in manually

data_frame_5 = data_frame_4.interpolate()
new_value = data_frame_5.iloc[1]['2018'] * .95
data_frame_5.at[0, '2018'] = new_value

new_value_2 = data_frame_5.iloc[1]['2019'] * .95
data_frame_5.at[0, '2019'] = new_value_2
```
